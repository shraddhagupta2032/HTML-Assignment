1.  Explain the basic structure of an HTML page. Why is it important to follow
this structure?

 - The basic HTML structure is like the skeleton of a webpage — it organizes content so browsers and users can understand it correctly. Without it, the page may not display or function properly.

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title></title>
</head>
<body> // This contain the visible part
    <h1>this tag is used for heading</h1>
    <p>This is for writing a paragraph.</p>
</body>
</html>
```

-  Its important to follow this basic structure to improve reliability.
As a clear structure makes it easier to read, debug and update your code.

4.  You’re creating a form that asks the user to enter their name, select their
country from a list, and click a button to submit. Write the HTML code for this.

```python

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Creating a form</title>
</head>
<body>
<form>
<label for="name">Name:</label>
<input type="text" id="name" name="name" required>
<br>
<label for="country">Country</lable>
<select>
<option value="India">India</option>
<option value="USA">USA</option>
<option value="UK">UK</option>
</select>
<br>
<button type="submit">Submit</button>
</form>
</body>
</html>
```

5. Write short HTML code to show your favorite fruits in a bulleted list and
